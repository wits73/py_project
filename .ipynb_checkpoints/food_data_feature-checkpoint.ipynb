```python
import numpy as np
import pandas as pd
```

```python
# make the countries string to lower case
df_base.countries = df_base.countries.str.lower()
```

```python
# extract only us
df_us_filtered = df_base[(df_base['countries'] == 'united states') | (df_base['countries'] == 'en:us') \
                         | (df_base['countries'] == 'us')]
```

```python
# extract columns
lst_extract_columns = [
    "code","created_datetime","product_name","brands", 
    "ingredients_text","allergens","serving_size","ingredients_from_palm_oil_n", "main_category",
    "energy_100g", "fat_100g", "saturated-fat_100g", "carbohydrates_100g", "sugars_100g", "salt_100g", "sodium_100g"
]

#"code","url", "created_datetime","product_name","brands","countries", 
#"ingredients_text","allergens","serving_size","additives","ingredients_from_palm_oil_n", "main_category",
#"energy_100g", "fat_100g", "saturated-fat_100g", "carbohydrates_100g", "sugars_100g", "fiber_100g", "proteins_100g",
#"salt_100g", "sodium_100g", "calcium_100g", "iron_100g","vitamin-d_100g", "vitamin-e_100g","vitamin-k_100g",
#"vitamin-c_100g","vitamin-b1_100g", "vitamin-b2_100g", "vitamin-pp_100g","vitamin-b6_100g","vitamin-b9_100g"
```

```python
df_us_column_filtered = df_us_filtered[lst_extract_columns]
```

```python
df_us_column_filtered['created_datetime']  = \
    df_us_column_filtered['created_datetime'].astype('datetime64[ns]').dt.strftime('%m/%d/%Y')
#pd.to_datetime(df_us_column_filtered['created_datetime'], format='%Y%m%d', errors='ignore')
```

```python
lst_ingridient_columns = ['ingredients_from_palm_oil_n','energy_100g','fat_100g','saturated-fat_100g','carbohydrates_100g' \
                          ,'sugars_100g','salt_100g','sodium_100g']
```

```python
lst_str_nan = ['product_name','brands','ingredients_text','allergens','serving_size','main_category']
```

```python
# remove space
df_us_rm_column_filtered['product_name'] = df_us_rm_column_filtered['product_name'].str.strip()
df_us_rm_column_filtered['brands'] = df_us_rm_column_filtered['brands'].str.strip()
df_us_rm_column_filtered['ingredients_text'] = df_us_rm_column_filtered['ingredients_text'].str.strip()
df_us_rm_column_filtered['allergens'] = df_us_rm_column_filtered['allergens'].str.strip()
df_us_rm_column_filtered['serving_size'] = df_us_rm_column_filtered['serving_size'].str.strip()
df_us_rm_column_filtered['main_category'] = df_us_rm_column_filtered['main_category'].str.strip()
```

```python
# NaN makes an error
df_us_rm_column_filtered[lst_str_nan] = \
df_us_rm_column_filtered[lst_str_nan].fillna("None").values
```
